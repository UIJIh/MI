```python
import fitz  # PyMuPDF
import csv

pdf_file = "9.pdf"
doc = fitz.open(pdf_file)

# idiom과 meaning 쌍을 저장할 리스트
idiom_meaning_pairs = []

# 모든 페이지를 순회 (필요에 따라 특정 페이지 범위를 지정할 수 있음)
for page in doc[9:12]:
    # 페이지의 텍스트 정보를 dict 형태로 가져옴
    blocks = page.get_text("dict")["blocks"]
    
    for block in blocks:
        if "lines" not in block:
            continue
        
        spans = []
        for line in block["lines"]:
            for span in line["spans"]:
                # 제외할 조건: 특정 글꼴 및 크기 조합
                if (
                    (span["size"] == 13.0 and 
                     span["font"] in ("MyriadArabic-Regular", "MyriadArabic-It") and 
                     20 <= span["origin"][1] <= 40)
                    or
                    (span["size"] == 30.0 and span["font"] == "MyriadArabic-Bold")
                ):
                    continue  # 제외 대상이면 리스트에 추가하지 않음
                
                spans.append(span)

        # 연속된 같은 font 그룹화
        grouped_spans = []
        temp_text = spans[0]["text"] if spans else ""
        temp_font = spans[0]["font"] if spans else ""

        for i in range(1, len(spans)):
            if spans[i]["font"] == temp_font:
                temp_text += " " + spans[i]["text"]
            else:
                grouped_spans.append({"font": temp_font, "text": temp_text.strip()})
                temp_text = spans[i]["text"]
                temp_font = spans[i]["font"]

        if temp_text:  # 마지막 그룹 추가
            grouped_spans.append({"font": temp_font, "text": temp_text.strip()})

        # idiom과 meaning을 찾아 저장
        i = 0
        while i < len(grouped_spans) - 1:
            if (grouped_spans[i]["font"] == "TimesNewRomanPSMT" and 
                grouped_spans[i + 1]["font"] == "MyriadArabic-Regular"):

                idiom = grouped_spans[i]["text"]
                meaning = grouped_spans[i + 1]["text"]
                
                if idiom and meaning:
                    idiom_meaning_pairs.append((idiom, meaning))

                i += 2  # 쌍을 찾았으므로 다음 쌍으로 이동
            else:
                i += 1

# CSV 파일로 저장 (UTF-8 BOM 포함)
with open("arabic.csv", "w", newline="", encoding="utf-8-sig") as csvfile:
    writer = csv.writer(csvfile)
    writer.writerow(["Idiom", "Meaning"])
    for pair in idiom_meaning_pairs:
        writer.writerow(pair)

print(f"총 {len(idiom_meaning_pairs)}개의 idiom-meaning 쌍이 추출되어 arabic.csv 파일에 저장되었습니다.")
```

```python
import pandas as pd
import deepl

# DeepL API 키 설정
auth_key = "914827b6-e548-49f8-a871-ab80580c0808:fx" 
translator = deepl.Translator(auth_key)

# CSV 파일 로드
df = pd.read_csv("arabic_4.csv")

# 두 번째 칼럼(영어 문장)을 번역
def translate_text(text, index):
    try:
        result = translator.translate_text(text, target_lang="AR")
        return result.text
    except Exception as e:
        print(f"Error translating row {index}: '{text}': {e}")
        return text  # 오류 발생 시 원래 텍스트 반환

# 새 칼럼 추가하여 번역 수행
df["Translated_Arabic"] = df.iloc[:, 1].astype(str).apply(lambda text, index: translate_text(text, index), index=df.index)

# 번역된 파일 저장
df.to_csv("arabic_5_trans.csv", index=False)

print("번역 완료!")
```
