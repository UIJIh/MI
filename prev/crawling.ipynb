```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time

# Initialize WebDriver
driver = webdriver.Chrome()

# URL of the webpage
url = 'https://matterhornlanguages.com/german-idioms/'  # Replace with the actual URL if different
driver.get(url)

# Handle the cookie consent banner
try:
    accept_button_xpath = '//*[@id="wt-cli-accept-all-btn"]'
    accept_button = WebDriverWait(driver, 10).until(
        EC.element_to_be_clickable((By.XPATH, accept_button_xpath))
    )
    accept_button.click()
    time.sleep(2)  # Wait for the banner to disappear
except Exception as e:
    print(f"No cookie banner to accept: {e}")

# Data storage
idioms_data = []

try:
    # Loop through the main accordion sections
    for i in range(1, 28):  # Assuming there are 6 main sections, adjust as needed
        try:
            # Expand the main accordion section
            main_section_xpath = f'//*[@id="heading{i}"]/button'
            main_button = WebDriverWait(driver, 10).until(
                EC.element_to_be_clickable((By.XPATH, main_section_xpath))
            )
            main_button.click()
            time.sleep(1)

            # Loop through the sub-accordion items within the main section
            sub_accordion_xpath = f'//div[@id="collapse{i}"]//div[contains(@class, "card")]'
            sub_accordions = driver.find_elements(By.XPATH, sub_accordion_xpath)

            for j in range(len(sub_accordions)):
                try:
                    # Expand the sub-accordion item
                    sub_button_xpath = f'//*[@id="subheading{i}-{j}"]/button'
                    sub_button = WebDriverWait(driver, 10).until(
                        EC.element_to_be_clickable((By.XPATH, sub_button_xpath))
                    )
                    sub_button.click()
                    time.sleep(1)

                    # Extract the data
                    german_idiom_xpath = f'//*[@id="subheading{i}-{j}"]/button'
                    german_idiom = driver.find_element(By.XPATH, german_idiom_xpath).text

                    literal_translation_xpath = f'//*[@id="subcollapse{i}-{j}"]//p[1]'
                    literal_translation = driver.find_element(By.XPATH, literal_translation_xpath).text

                    english_translation_xpath = f'//*[@id="subcollapse{i}-{j}"]//p[2]/strong'
                    english_translation = driver.find_element(By.XPATH, english_translation_xpath).text

                    # Store the extracted data
                    idioms_data.append({
                        'german_idiom': german_idiom,
                        'literal_translation': literal_translation,
                        'english_translation': english_translation
                    })

                    print(f"Extracted data from section {i}, item {j}")
                    time.sleep(0.5)
                except Exception as sub_item_exception:
                    print(f"Error processing sub-item {j} in section {i}: {sub_item_exception}")
                    continue

        except Exception as main_section_exception:
            print(f"Error processing main section {i}: {main_section_exception}")
            continue

finally:
    driver.quit()

# Output the results
for idiom in idioms_data:
    print(f"German Idiom: {idiom['german_idiom']}")
    print(f"Literal Translation: {idiom['literal_translation']}")
    print(f"English Translation: {idiom['english_translation']}\n")
```
